```python
import pandas as pd
import matplotlib.pyplot as plt

# Set the matplotlib backend to inline
%matplotlib inline

# Read the CSV file
data = pd.read_csv('wret1.csv')

wret = data.iloc[:, 0].tolist()
exec = data.iloc[:, 1].tolist()

error = []
for i in range(1, len(exec)):
	error.append((wret[i - 1] / exec[i] - 1) * 100)

datao = pd.read_csv('wret1o.csv')

wret1 = datao.iloc[:, 0].tolist()
exec1 = datao.iloc[:, 1].tolist()

wret2 = datao.iloc[:, 2].tolist()
exec2 = datao.iloc[:, 3].tolist()

wret3 = datao.iloc[:, 4].tolist()
exec3 = datao.iloc[:, 5].tolist()

wret4 = datao.iloc[:, 6].tolist()
exec4 = datao.iloc[:, 7].tolist()

wret = wret[:-1]
exec = exec[1:]

wret1 = wret1[:-1]
exec1 = exec1[1:]

wret2 = wret2[:-1]
exec2 = exec2[1:]

wret3 = wret3[:-1]
exec3 = exec3[1:]

wret4 = wret4[:-1]
exec4 = exec4[1:]

# Set the width of the plot
plt.figure(figsize=(20, 10))  # Adjust width as needed

# plt.plot(error)

plt.plot(wret, color='blue', linestyle='dotted', linewidth=1.5)
plt.plot(exec, color='blue', linestyle='solid', linewidth=1)

# plt.plot(wret1)
# plt.plot(exec1)

# plt.plot(wret2)
# plt.plot(exec2)

# plt.plot(wret3)
# plt.plot(exec3)

plt.plot(wret4, color='red', linestyle='dotted', linewidth=1.5)
plt.plot(exec4,  color='red', linestyle='solid', linewidth=1)

plt.legend(['wret', 'exec', 'wret1', 'exec1', 'wret2', 'exec2', 'wret3', 'exec3', 'wret4', 'exec4'], loc='upper left')

plt.title('Data from wret.csv')
plt.show()

average_error = sum(error) / len(error)
print(average_error)
```
